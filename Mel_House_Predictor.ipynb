```python
import pandas as pd
import numpy as np
```

```python
mel_data = pd.read_csv('melb_data.csv')
mel_data.head()
```

```python
# From our data abouve we can see that there are some missing values so we have to delete them 
# We do this by dropping the Na values from th dataset 
# First we copy the data to another part

melborn_data = mel_data.copy()
melborn_data.head()
```

```python
melborn_data = melborn_data.dropna(axis = 'rows')
```

```python
# Now lets check if the rows containg NaN values have been dropped 
melborn_data.head()
```

```python
# Lets get the target and put it into a varable called y 
y = melborn_data['Price']
```

```python
#and Lets take some important and very relevant features as what our model will train with 

melborn_features = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']
```

```python
# Lets fix this into the variable X for easy calling 

X = melborn_data[melborn_features]
```

```python
# Great Now lets check our data one by one 
X.head()
```

```python
X.describe()
```

```python
X.info()
```

 ## Training the Model 

```python
# Lets import the Sikit Learn Libraray that will help us train the model 
# We will be using the Sklearn Decision Tree for traing our model 

from sklearn.tree import DecisionTreeRegressor 

# Lets define our model and speciofie the number of random state fro us to always get same result in each run 

melborn_model = DecisionTreeRegressor(random_state = 1)
```

```python
# Time to Train our model baby 

melborn_model.fit(X, y)
```

## Great Our model has beeen trained time to do some predictions 

```python
print('Making predictions for the first five houses : \n ')
print(X.head(), '\n')
print("The Predictions are : \n ")
print(melborn_model.predict(X.head()), 'Prices in Dollars')
```
